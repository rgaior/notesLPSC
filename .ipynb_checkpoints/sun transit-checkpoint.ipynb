##quiet sun

The quiet sun component is parameterized between 1 and 20GHz with:
\begin{equation}
    Sq \ [SFU] = 26.4 + 12.4 \ f + 1.11\ f^2 \\ \rm for  (1 < f(GHz) < 20) 
\end{equation}
with f in GHz

```python
%matplotlib inline
```

```python
f = np.linspace(0.8,5,10)
sq = 26.4 + 12.4*f + 1.11*f**2 
plt.plot(f,sq)
plt.xlabel('frequency [GHz]')
plt.ylabel('Quiet sun flux [SFU]')
plt.grid()
```

## slowly varying component

This component is made of two components, the 11 years cycle and a ~27 days cycle.  
The value of this component with frequency can be parameterized with the value at 2.8GHz (a.k.a F10.7. the flux at 10.7cm)

The parameterization function is:
\begin{equation}
    Sv  \ [SFU]= \frac{0.64 ( F10.7 - 70 ) f^{0.4}}{ 1 + 1.56 ( ln \ ( \ f  \ / \  2.9 ) )^2}
\end{equation}
with f in GHz. The value 70 is the quiet sun at 2.8GHz  
The dependence in frequency is evaluated taking (F10.7 - 70 = 1):

```python
f = np.linspace(0.8,5,100)
sv = (0.64*f**0.4)/(1+ 1.56*(np.log(f/2.9))**2)
plt.plot(f,sv)
plt.xlabel('frequency [GHz]')
plt.ylabel('Slow varying sun flux [SFU]')
plt.grid()
```

**remark** it is not exactly 1 at 2.8GHz ! Don't know really why...

**other remark:** there is another component which is the burst component. This is the transient and can increase the flux by several orders of magnitude on the scale of minute to hours. We need to take care that the measurement can be reproducible on several days because if we observe juste once the sun, we could be in a burst.

We can then obtain the expected flux or the temperature for a certain day as a function of the hour with:
\begin{equation}
    P (\theta, \phi) = \frac{1}{2}\cdot F_{sun}(\theta,\phi) \cdot A_{eff}(\theta,\phi) \cdot \Delta \nu
\end{equation}

ou 
\begin{equation}
    T(\theta,\phi) = \frac{\frac{1}{2}\cdot F_{sun}(\theta,\phi) \cdot A_{eff}(\theta,\phi)}{k_{B}}
\end{equation}




```python
import os
os.chdir('/Users/romain/work/Auger/EASIER/LPSC/suntransit/script/test/')
%run testdailyflux.py 2014 12 21
```

Or if we set a system temperature we can obtain the raise in dB with:
\begin{equation}
    \Delta dB = 10.log_{10}\frac{T_{sys} + T_{sun}}{T_{sys}}
\end{equation}

and 

\begin{equation}
    \Delta ADC = 50\cdot \Delta dB
\end{equation}
