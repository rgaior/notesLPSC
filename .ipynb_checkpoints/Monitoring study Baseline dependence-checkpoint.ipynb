#intro

The aim of this study is to understand the variation of the baseline for the EASIER antennas.
This is especially helpful for the sun transit measurement (or non measurement).  
From Corinne studies, the sun can be observed with the Norsat antennas, meaning that the system temperature can be measured in situ. For the EASIER antenna, the sun transit is expected to raise the level of the baseline by around 10 ADC in the best case,i.e. low system temperature. (EASIER antennas have a wider lobe, that is to say the sun occupies a smaller solid angle, thus the raise in temperature given by the sun transit is lower.)  
However, we know that our detector has a dependence in temperature, and also for the EASIER 61 antennas, a dependence in supplied voltage.  
We want here to study how much the baseline variations can be corrected so that the sun transit can be studied.
